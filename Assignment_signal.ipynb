```python
import numpy as np
import librosa
import matplotlib.pyplot as plt


audio_path = 'voice.aac'
y, sr = librosa.load(audio_path, sr=None)

# Display the audio signal
plt.figure(figsize=(10, 5))
plt.plot(y)
plt.title('Audio Signal')
plt.xlabel('Sample Number')
plt.ylabel('Amplitude')
plt.show()
```

```python
# Compute the FFT of the signal
y_fft = np.fft.fft(y)
N = len(y)

# Frequency bins
frequencies = np.fft.fftfreq(N, d=1/sr)

# Magnitude of the FFT (absolute value)
magnitude = np.abs(y_fft) / N

# Only take the positive frequencies (symmetry)
positive_frequencies = frequencies[:N//2]
positive_magnitude = magnitude[:N//2]

# Plot the frequency-domain signal
plt.figure(figsize=(10, 5))
plt.plot(positive_frequencies, positive_magnitude)
plt.title('Frequency-Domain Signal (Magnitude Spectrum)')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid()
plt.show()
```
